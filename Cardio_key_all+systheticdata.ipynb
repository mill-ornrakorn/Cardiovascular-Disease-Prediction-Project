```python
# standardizing
from sklearn import preprocessing

std_scaler = preprocessing.StandardScaler()
cardio_feature = pd.DataFrame(std_scaler.fit_transform(cardio_feature), columns = cardio_feature.columns)
cardio_feature
```

# Splitting data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cardio_feature, cardio_result, test_size=0.3)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
Logres_classifier = LogisticRegression(solver='lbfgs', max_iter=10000)
Logres_classifier.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
Logres_classifier2 = LogisticRegression(solver='newton-cg', max_iter=10000)
Logres_classifier2.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression

Logres_classifier3 =LogisticRegression(solver='lbfgs', max_iter=10000)
Logres_classifier3.fit(x_train,y_train)

acc = round(Logres_classifier3.score(x_train,y_train) * 100, 4)
acc_test = round(Logres_classifier3.score(x_test, y_test)* 100, 4)
print(acc, acc_test)
```

```python
from sklearn.linear_model import LogisticRegression

Logres_classifier4 =LogisticRegression(solver='saga', max_iter=10000)
Logres_classifier4.fit(x_train,y_train)

acc = round(Logres_classifier4.score(x_train,y_train) * 100, 4)
acc_test = round(Logres_classifier4.score(x_test, y_test)* 100, 4)
print('Accuracy_train = ',acc ,'Accuracy_test = ',acc_test)
```
